## Data Cleaning

#### Code

```python
# rename the column 'A' to 'new_name'
df_pxn = df_pxn.rename(columns={'Value': 'Value (mt)'})  # mt means Metric tonnes

df_sup = df_sup.rename(columns={'Value': 'Value (kcal/person/day)'})
```

```python
# remove the closing parenthesis from the Year column
df_pxn['Year'] = df_pxn['Year'].str.replace(')', '')

# convert the Year column to an integer data type
df_pxn['Year'] = df_pxn['Year'].astype(int)
```

```python
df_pxn['Item'] = df_pxn['Item'].astype('category')

df_pxn['Country'] = df_pxn['Country'].astype('category')
df_sup['Country'] = df_sup['Country'].astype('category')
```

```python
sns.set_theme(style='darkgrid') 
base_colour = sns.color_palette()[-1]
```

```python
df_sup.hist('Value (kcal/person/day)', color=base_colour);
plt.title('Histogram of Supply Values');
```

```python
plt.figure(figsize=(17,3))
dots = dict(markerfacecolor=base_colour, marker='o')
plt.boxplot(df_pxn['Value (mt)'], vert=False, flierprops = dots);
plt.yticks([],[]);
plt.xticks(fontsize=13)
plt.title('Distribution of Production Values', fontsize=15);
```

It appears that majority of the values are between 0 and 5000. A scaled Histogram and zoomed boxplot will do more justice to visualizing the data.

```python
plt.figure(figsize=(15,7))

log_max = np.log10(df_pxn['Value (mt)'].max())
bins=10**np.arange(0,log_max+0.5,0.5)

# Generate the x-ticks to apply
ticks = [0.1, 0.3, 1, 3, 10, 30, 100, 300, 1000, 3000, 10000, 30000, 55000]
# Convert ticks into string values, to be displayed along the x-axis
labels = ['{}'.format(v) for v in ticks]

plt.hist(df_pxn['Value (mt)'], bins = bins, color=base_colour);
plt.xscale('log')
plt.xticks(ticks, labels);
plt.xlim(0.9,55000);
plt.title('Scaled histogram of Production Values', fontsize=15);
```

```python
plt.figure(figsize=(17,3))
dots = dict(markerfacecolor=base_colour, marker='o')
plt.boxplot(df_pxn['Value (mt)'], vert=False, flierprops = dots);
plt.yticks([],[]);
plt.xticks(fontsize=13)
plt.xlim(0,400)
plt.title('Distribution of Value', fontsize=15);
```

Now, we see that for the amount of food produced, Majority of the data is spread between 0 metric tonnes  and about 260 metric tonnes, with outliers reaching values of up to 50,000 metric tonnes.

```python
plt.figure(figsize=(15,7))
top_items = df_pxn['Item'].value_counts().head(10)
sns.countplot(data=df_pxn, x='Item', order=top_items.index, color =base_colour);
plt.xticks(rotation=30)
plt.title('Top 10 produced food items by frequency', fontsize=15)
plt.xlabel('Item', fontsize=13)
plt.ylabel('');
```

```python
plt.figure(figsize = (10,7))


# get the top 10 countries by total food production
top_countries = df_pxn.groupby('Country')['Value (mt)'].sum().sort_values(ascending=False)[:10].index.tolist()

# filter the data to only include the top 10 countries
df_top10 = df_pxn[df_pxn['Country'].isin(top_countries)]

# plot a bar plot of the top 10 countries by food production
sns.barplot(x='Country', y='Value (mt)', data=df_top10, order=top_countries, color = base_colour)
plt.xlabel('Country')
plt.ylabel('Weight (metric tonnes)')
plt.title('Top 10 African Countries by Food Production (2004-2013)')
plt.xticks(rotation = 15)
plt.show()
```

## Multivariate Exploration

To compare Rice production vs rice supply in Nigeria.

```python
food_item = 'Rice (Milled Equivalent)'
food_item_2 = 'Cereals, Other'
country = 'Nigeria'

# filter the data to only include the specified food item and country
df_filtered = df_pxn[(df_pxn['Item'] == food_item) & (df_pxn['Country'] == country)]
df_filtered_2 = df_pxn[(df_pxn['Item'] == food_item_2) & (df_pxn['Country'] == country)]


plt.figure(figsize=(13,7))
plt.suptitle('Comparing Rice production to Other cereals', fontsize=15)

# plot a scatter plot of food production over time
sns.lineplot(x='Year', y='Value (mt)', data=df_filtered, color=base_colour, label=food_item)
sns.lineplot(x='Year', y='Value (mt)', data=df_filtered_2, color=sns.color_palette("dark")[-1], label=food_item_2)

plt.xlabel('Year')
plt.ylabel('Weight (metric tonnes)')
plt.title(f'{food_item} and {food_item_2} Production in {country} (2004-2013)')
plt.legend()

plt.show()
```

Average Calories consumed per day by Countries.

```python
plt.figure(figsize=(15,9))

# Selecting the top 10 countries by mean 'Value (kcal/person/day)'
top_10 = df_sup.groupby('Country')['Value (kcal/person/day)'].mean().sort_values(ascending=False).head(10).index.tolist()

# Creating a subset of the dataframe with only the top 10 countries
df_top_10 = df_sup[df_sup['Country'].isin(top_10)]

# Creating a scatterplot with hue='Country' and the custom palette
sns.lineplot(data=df_top_10, x='Year', y='Value (kcal/person/day)', hue='Country', color=base_colour)

# Removing the countries not in the top 10 from the legend
handles, labels = plt.gca().get_legend_handles_labels()
labels_filtered = [label for label in labels if label in top_10]
handles_filtered = [handles[labels.index(label)] for label in labels_filtered]
plt.gca().legend(handles_filtered, labels_filtered)

# Setting the plot title and axes labels
plt.title("Average Calories Consumed per Person per Day (Top 10 Countries)")
plt.xlabel("Year")
plt.ylabel("Calories per Person per Day")

# Displaying the plot
plt.show()
```

### For Quiz

```python
# Filter the dataset to only include rice data
rice_data = df_pxn[df_pxn['Item'] == 'Rice (Milled Equivalent)']

# Group the rice data by country and sum the quantity produced from 2004 to 2013
rice_by_country = rice_data[rice_data['Year'].between(2004, 2013)].groupby('Country')['Value (mt)'].sum()

# Sort the grouped data in descending order by the total quantity of rice produced
rice_by_country = rice_by_country.sort_values(ascending=False)

# Select the top three countries from the sorted data
top_three = rice_by_country.head(3)
top_three
```

```python
# Filter the dataset to only include data for the years 2004 to 2006
food_data = df_sup[df_sup['Year'].between(2004, 2006)]

# Group the food data by country and calculate the total food supply during the specified time period
food_by_country = food_data.groupby('Country')['Value (kcal/person/day)'].sum()

# Sort the grouped data in ascending order by the total food supply
food_by_country = food_by_country.sort_values()

# Select the bottom three countries from the sorted data
bottom_three = food_by_country.head(3)
bottom_three
```

```python
# Filter the dataset to only include data for the year 2010 and the food item "Oats"
oats_data = df_pxn[(df_pxn['Year'] == 2010) & (df_pxn['Item'] == 'Oats')]

# Group the oats data by country and calculate the total quantity of oats produced in 2010
oats_by_country = oats_data.groupby('Country')['Value (mt)'].sum()

# Sort the grouped data in descending order by the total quantity of oats produced
oats_by_country = oats_by_country.sort_values(ascending=False)

# Select the top country from the sorted data
top_country = oats_by_country.index[0]

# Print the top country and the total quantity of oats produced in 2010
print(f"The country that produced the highest amount of oats in 2010 is {top_country} with a total production of {oats_by_country[0]} metric tons.")
```
